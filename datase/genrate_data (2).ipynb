```python
system_prompt = """
You are a language model trained to generate diverse and realistic todo lists for training a machine learning model. When generating todo lists, please adhere to the following guidelines:

1. **Placeholder Use:**
   - Replace any specific times mentioned with "{time}".
   - Replace any person names with "{name}".

2. **List Content:**
   - Each todo list should detail at least one of the following: time, place, participants, or purpose. The format should mimic real todo lists written by humans.

3. **Output Structure:**
   - Generate multiple todo lists in a single response. Separate each todo list with the delimiter "||".

4. **Diversity:**
   - Ensure diversity in the types of todo lists to cover various contexts and scenarios, enhancing the machine learning model's effectiveness.

Example:

- Buy groceries at {name}'s store at {time} || Attend the weekly meeting online at {time} with {name} and {name} || Schedule a car service appointment at the nearest garage || Plan a birthday party for {name} at {time} at the local park
"""
```

```python
completion = client.chat.completions.create(
  model="gpt-4",
  messages=[
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": "please genrate 30 to do list"}
  ],
  temperature=0.9
)

result = completion.choices[0].message
```
